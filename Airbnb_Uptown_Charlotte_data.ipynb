```python
# Import the required libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Using the read_csv function and Path module, created a DataFrame 
# by importing uptown_charlotte_rental_data.csv file 
rental_data_df = pd.read_csv(Path("uptown_charlotte_rental_data.csv")).set_index('Apartment Complex')

#Review the first and last five rows of the DataFrame
display(rental_data_df)
```

```python
# Load neighborhoods coordinates data
complex_locations_df = pd.read_csv(Path("apartment_coordinates.csv")).set_index('Apartment Complex')

# Review the DataFrame
display(complex_locations_df)   
```

```python
# Using the Pandas `concat` function, joined the 
# neighborhood_locations_df and the apartment_list_df DataFrame

apartment_list_df = pd.merge(
    rental_data_df, complex_locations_df,
    left_index= True,
    right_index=True,
    how= 'left',
    sort=False,
    
)
    
# Review the resulting DataFrame
display(apartment_list_df)
```

```python
# Call the dropna function to remove any neighborhoods that do not have data
#apartment_list_df = apartment_list_df.reset_index().dropna()

#Rename the "index" column as "Apartment Complex" for use in the Visualization
apartment_list_df=apartment_list_df.rename(columns={"index": "Apartment Complex"})

# Review the resulting DataFrame
display(apartment_list_df)
```

```python
# Created a plot to analyze neighborhood info
apartment_list_df.hvplot.points(
    x='Lon',
    xlabel= "Longitude",
    y='Lat',
    ylabel= "Latitude",
    geo=True,
    size= 'SQ.FT',
    color= 'Price (mo)',
    frame_width= 700,
    frame_height= 500,
    title= "Uptown Charlotte Available Rental Properties",
    tiles= True,
    hover=True,
    hover_cols=['Unit Type','Apartment Complex','Deposit', 'Apt Number', 'Availabillity'],
    clabel='Rental Price per Month'
)
```

```python
# calculating average rents per apartment complex
avg_rents = apartment_list_df.groupby('Apartment Complex').mean().reset_index()[['Apartment Complex', 'Rent']]
avg_rents['Rent'] = [round(r) for r in avg_rents.Rent] 
avg_rents.sort_values(by=['Rent'], ascending=False, inplace=True)
```
